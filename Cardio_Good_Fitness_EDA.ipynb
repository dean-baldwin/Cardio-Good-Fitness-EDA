### Create DataFrames from the 3 datasets

```python
cardio_df = pd.DataFrame(data)
```

#### Create seperate DataFrames for each product

```python
tm195 = cardio_df[cardio_df['Product'] == 'TM195']
tm498 = cardio_df[cardio_df['Product'] == 'TM498']
tm798 = cardio_df[cardio_df['Product'] == 'TM798']
```

## Review of Products by Age
Is there a variation in product selection based on the age of the customer?

```python
sns.displot(cardio_df['Age'], kde = True, rug = True);
```

 **Observations:**
 1. The median age of customer within the dataset is ~25 years old
 2. There appears to be a positively skewed (right-skewed) distribution in regards to age

```python
sns.boxplot(x=cardio_df['Product'], y=cardio_df['Age']);
```

 **Observations:**
 1. TM195 and TM498 are very similar in regards to age
 2. Customers who purchased the TM798 appear to be a smaller more focused subset of the age range associated with the other two products.
 3. The TM798 has a larger number of outliers
 4. The median age of customer for all products is relatively the same
